# N-dimension

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df_original = pd.read_csv("Diabetes.csv")
df_original.head()
```

```python
cols = [c for c in df_original.columns if c not in ["Pregnancies", "Outcome"]]
df = df_original.copy()
df[cols] = df[cols].replace({0: np.NaN})
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Scatter Matrix

```python
pd.plotting.scatter_matrix(df, figsize=(10, 10));
```

```python
df2 = df.dropna()
colors = df2["Outcome"].map(lambda x: "blue" if x else "red")
pd.plotting.scatter_matrix(df2, figsize=(10, 10), c=colors);
```

## Correlation Plots
